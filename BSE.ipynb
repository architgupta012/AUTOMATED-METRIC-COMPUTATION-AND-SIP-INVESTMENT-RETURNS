### Calculating Daily Returns for each index
    Daily returns for a date is the ratio of closing price for that date to the closing price to the previous date.

```python
# To get the Index names in a list called lst
lst=[]
lst= df1.columns
lst=lst[1:]
```

```python
## df1 daily returns 

for x in lst:
    y=x[:-7]
    z=y+" DAILY RETURN"
    df1[z]=0.0
    for i in range(1,len(df1["DATE"])):
        df1[z][i]= ln(df1[x][i-1]/df1[x][i])*100
```

```python
#enter the date from where the metrics are to be calculated
start_date= "2022-05-22"
format = '%Y-%m-%d'
start_date = datetime.datetime.strptime(start_date,format)

yr= start_date.year
yr1= start_date.year
mn= start_date.month
dy= start_date.day


def Nearest(items, pivot): 
    if pivot in items:
        return pivot
    else:
        return min(items, key=lambda x: abs(pivot-x)) 


#getting the nearest date

d1= start_date
d1=Nearest(df1["DATE"][(df1['DATE'] <= d1)],d1)
yr=yr-1
d2= datetime.datetime(yr,mn,dy)
d2= Nearest(df1["DATE"][(df1['DATE'] <= d2)],d2)
yr=yr-1
d3= datetime.datetime(yr,mn,dy)
d3= Nearest(df1["DATE"][(df1['DATE'] <= d3)],d3)
yr=yr-1
d4= datetime.datetime(yr,mn,dy)
d4= Nearest(df1["DATE"][(df1['DATE'] <= d4)],d4)
if yr1>2016:
    yr=yr-2
    d5= datetime.datetime(yr,mn,dy)
    d5= Nearest(df1["DATE"][(df1['DATE'] <= d5)],d5)
if yr1>2021:
    yr=yr-5
    d6= datetime.datetime(yr,mn,dy)
    d6= Nearest(df1["DATE"][(df1['DATE'] <= d6)],d6)
    
    
df2=df1.loc[df["DATE"]==d1.strftime(format)]
df3=df1.loc[df["DATE"]==d2.strftime(format)]
df4=df1.loc[df["DATE"]==d3.strftime(format)]
df5=df1.loc[df["DATE"]==d4.strftime(format)]
df6=df1.loc[df["DATE"]==d5.strftime(format)]
df7=df1.loc[df["DATE"]==d6.strftime(format)]
 
df2=df2.append(df3)
df2=df2.append(df4)
df2=df2.append(df5)
if yr1>2016:
    df2=df2.append(df6)
if yr1>2021:
    df2=df2.append(df7)
df2= df2.reset_index()
df2= df2.drop_duplicates()
df2= df2.drop("index",axis=1)
df2
```

### Calculating Yearly Returns  for the entered date
    formula for yearly return used:
    for dates which are 'p' years apart and index prices for date1 is 'x' and date2 is 'y', the formula will be:
    return= (((x/y)^(1/p))-1)*100

```python
df_yr_ret= pd.DataFrame()

if yr1>2021:    
    df_yr_ret["YEARS"]=["1 Year","2 Years","3 Years","5 Years","10 Years"]
elif yr1> 2016:
    df_yr_ret["YEARS"]=["1 Year","2 Years","3 Years","5 Years"]
else:
    df_yr_ret["YEARS"]=["1 Year","2 Years","3 Years"]



for y in lst:
    x= y[:-7]
    
    df_yr_ret[x]=0.0
    calc=[]
    for i in range(0,len(df2["DATE"])-1):
        yr= df2["DATE"][0].year-df2["DATE"][i+1].year
        pw=1/yr
        bs= (df2[y][0])/(df2[y][i+1])
        cal=((pow(bs,pw)-1)*100)
        calc.append(cal)  
    df_yr_ret[x]=calc
df_yr_ret   
```

### Calculating Annual Volatility
     formula used: annual volatility= (sqrt(no.of days in an year) * standard deviation for the period)
     

```python
#yearly annual volatility 

#EDIT HERE IF YOU WANT TO CHANGE NO.OF DAYS IN AN YEAR
days=365  # days taken in an year 

df_yr_vol= pd.DataFrame()

if yr1>2021:    
    df_yr_vol["YEARS"]=["1 Year","2 Years","3 Years","5 Years","10 Years"]
elif yr1> 2016:
    df_yr_vol["YEARS"]=["1 Year","2 Years","3 Years","5 Years"]
else:
    df_yr_vol["YEARS"]=["1 Year","2 Years","3 Years"]

for y in lst:
    x = y[:-7]
    df_yr_vol[x]=0.0
    for i in range(0,len(df2["DATE"])-1):
        b= math.sqrt(days)*df_yr_std[x][i]
        df_yr_vol[x][i]=b
df_yr_vol
```

```python
invest_date="2012-12-21"
format = '%Y-%m-%d'
invest_date = datetime.datetime.strptime(invest_date,format)


yr= invest_date.year
yr1= invest_date.year
mn= invest_date.month
dy= invest_date.day

def Nearest(items, pivot):
    if pivot in items:
        return pivot
    else:
        return min(items, key=lambda x: abs(pivot-x)) 
    

    
df_sip= pd.DataFrame()
monthly_dates=[]
monthly_dates.append(datetime.datetime(yr,mn,dy))
next_date= invest_date
for i in range(0,no_mn-1):
    next_date = next_date.date() + relativedelta.relativedelta(months=1)
    nd=next_date.day
    nm=next_date.month
    ny=next_date.year
    n_date = datetime.datetime(ny,nm,nd)
    next_date=Nearest(df["DATE"][(df['DATE'] >= n_date)],n_date)
    monthly_dates.append(next_date)
    nm=next_date.month
    ny=next_date.year
    next_date= datetime.datetime(ny,nm,dy)
df_sip["DATE"]=monthly_dates
```
